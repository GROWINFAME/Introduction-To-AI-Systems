## **Вектор**


> Вектор представляет собой массив скаляров (минимально определяется двумя скалярами), по сути, описывающих координаты какой-то точки в конечномерном пространстве. Записываются они следующим образом: $x^⃗ \in ℝ^{n}$ (элемент n-мерного пространства действительных чисел)

\begin{align}
        x^⃗ = \begin{pmatrix}
        x_{11} \\
        x_{21} \\
        ...    \\
        x_{n1} \\
        \end{pmatrix}.
    \end{align}

или
$x^⃗=[x_1, x_2, ... x_n]$

> В компьютерных науках часто вектор представляет собой ***описание какой-то сущности***.

**Операции над векторами**

> Разность векторов (substraction): $ a^⃗ - b^⃗ = [3, 2] - [-2, 0] = [3+2, 2-0] = [5, 2]$



```python
vec_a = torch.tensor([3, 2])
vec_b = torch.tensor([-2, 0])

vec_a - vec_b

torch.subtract(vec_a, vec_b)

np.subtract(vec_a, vec_b)
```

> Сложение векторов (addition): $a^⃗ + b^⃗ = [3, 2] + [-2, 0] = [3-2, 2+0] = [1, 2]$



```python
vec_a + vec_b

torch.add(vec_a, vec_b)

np.add(vec_a, vec_b)
```

> Векторное поэлементное произведение - строка на столбец (произведения Адамара -  ***element-wise product*** или ***Hadamard product***): $[a^⃗ 𝗑 b^⃗] = [3, 2] 𝗑 [-2, 0] = [3*-2, 2*0] = [-6, 0]$

При умножении вектора на скаляр, произведение применяется аналогично поэлементно



```python
vec_a * vec_b

torch.multiply(vec_a, vec_b)

np.multiply(vec_a, vec_b)
```

> Скалярное произведение векторов (внутренее произведение - ***dot product, inner product*** или ***scalar product***) — это сумма произведение их компонентов, возвращающая скаляр. СТРОКА НА СТОЛБЕЦ.

$a^⃗ ⋅ b^⃗$ = [3,2] ⋅ [-2,0]' = 3⋅(-2) + 2⋅0 = -6

Геометрически оно определяется как произведение длины одного вектора на длину другого вектора и на косинус угла между ними.

$a^⃗ ⋅ b^⃗ = |a||b|cos(a)$

По сути оно отображает некую схожесть между двумя векторами в виде одного числа. 

При нормализованных векторах:
1.   Если скалярное произведение равно 0, то они находятся под углом 90 градусов и их ничего не связывает. 
2.   Если скалярное произведение положительное, то имеют одинаковое направление.
3.   Если скалярное произведение отрицательное, они отдаляются друг от друга.
4.   Если скалярное произведение векторов равно 1, они полностью совпадают.




```python
plt.figure(figsize=(10, 10))
plt.quiver([0], [0], [0], [1], color='r', units='xy', scale=1, angles='xy', scale_units='xy')
plt.quiver([0], [0], [1], [0], color='r', units='xy', scale=1, angles='xy', scale_units='xy')
plt.quiver([0], [0], [-1], [0], color='r', units='xy', scale=1, angles='xy', scale_units='xy')
plt.quiver([0], [0], [0], [-1], color='r', units='xy', scale=1, angles='xy', scale_units='xy')
plt.quiver([0], [0], [0.8], [0.5], color='b', units='xy', scale=1, angles='xy', scale_units='xy')
plt.quiver([0], [0], [0.7], [0.8], color='g', units='xy', scale=1, angles='xy', scale_units='xy')
plt.quiver([0], [0], [0.4], [0.9], color='black', units='xy', scale=1, angles='xy', scale_units='xy')
plt.xlim(-1, 1) # ограничения по x
plt.ylim(-1, 1) # ограничения по y
plt.grid() # построить сетку
plt.show()
```

> Поэлементное деление векторов (division)



```python
vec_a / vec_b

torch.divide(vec_a, vec_b)

np.divide(vec_a, vec_b)
```

## **Матрицы**



> Матрица — это массив скаляров одинаковой размерности. Обозначается жирным шрифтом в верхнем регистре. Матрица из m строк и n столбцов, записывается вот так: $A \in ℝ^{mхn}$

\begin{align}
        A = \begin{pmatrix}
        a_{11} & a_{12} & ... & a_{1n} \\
        a_{21} & a_{22} & ... & a_{2n} \\
        ... & ... & ... & ... \\
        a_{m1} & a_{m2} & ... & a_{mn} \\
        \end{pmatrix}.
    \end{align}

```python
data = torch.tensor([[70, 170, 40, 1],
                     [65, 169, 26, 0],
                     [77, 183, 19, 0]]) # например матрица с данными о весе, росте, возрасте и наличия хронической болезни конкретных людей
data[1, 0] # i - строка, j- столбец
```

```python
# Матрицы 2х2
A = torch.tensor([[1, 2], [3, 4]])
B = torch.tensor([[5, 6], [0, 7]])
```

**Операции над матрицами**

> Матричная разность (поэлементная)

\begin{align}
        A = \begin{pmatrix}
        1 & 2\\
        3 & 4\\
        \end{pmatrix} -
        \begin{pmatrix}
        5 & 6\\
        0 & 7\\
        \end{pmatrix}=
        \begin{pmatrix}
        1-5 & 2-6\\
        3-0 & 4-7 \\
        \end{pmatrix}=
         \begin{pmatrix}
        -4 & -4\\
        -3 & -3\\
        \end{pmatrix}
    \end{align}


```python
A - B
torch.subtract(A, B)
np.subtract(A, B)
```

> Матричное сложение (поэлементное)

\begin{align}
        A = \begin{pmatrix}
        1 & 2\\
        3 & 4\\
        \end{pmatrix} +
        \begin{pmatrix}
        5 & 6\\
        0 & 7\\
        \end{pmatrix}=
        \begin{pmatrix}
        1+5 & 2+6\\
        3+0 & 4+7 \\
        \end{pmatrix}=
         \begin{pmatrix}
        6 & 8\\
        3 & 11\\
        \end{pmatrix}
    \end{align}



```python
A + B
torch.add(A, B)
np.add(A, B)
```

> Произведение матрица (поэлементное)
\begin{align}
        A = \begin{pmatrix}
        1 & 2\\
        3 & 4\\
        \end{pmatrix} .
        \begin{pmatrix}
        5 & 6\\
        0 & 7\\
        \end{pmatrix}=
        \begin{pmatrix}
        1*5 & 2*6\\
        3*0 & 4*7 \\
        \end{pmatrix}=
         \begin{pmatrix}
        5 & 12\\
        0 & 28\\
        \end{pmatrix}
    \end{align}


```python
A * B
torch.multiply(A, B)
np.multiply(A, B)
```

> Матричное умножение: умножается каждый элемент строки на каждый элемент столбца после чего вычисляется их сумма:

\begin{align}
        A = \begin{pmatrix}
        1 & 2\\
        3 & 4\\
        \end{pmatrix} Х 
        \begin{pmatrix}
        5 & 6\\
        0 & 7\\
        \end{pmatrix}=
        \begin{pmatrix}
        1*5 + 2*0 & 1*6+2*7\\
        3*5 + 4*0 & 3*6+4*7 \\
        \end{pmatrix}=
         \begin{pmatrix}
        5 & 20\\
        15 & 46\\
        \end{pmatrix}
    \end{align}

Как мы можем заметить здесь осуществляется скалярное произведение вектров матриц.

Важно! При матричном умножении A должна иметь размер $nхm$ в то время как матрица B должна иметь размер $mхn$ 



```python
A = torch.tensor([[1, 2, 3,], [4, 5, 6,]]) # матрица 2х3
B = torch.tensor([[1, 1, 1], [2, 2, 2]]).T # транспонированная матрица 3х2

A @ B
torch.matmul(A, B)
A.mm(B)
np.matmul(A, B)
np.dot(A, B) # dot от Numpy в целом может работать как с векторами, так и с матрицами. Однако torch.dot работает только с 1D!
```

# **Функция и производная**

В более ясном языке, функция - это конструкция, которая хранит в себе зависимость одной величины от другой. Она принимает что-то на входе и возвращает что-то на выходе по определённому правилу.


```python
def linear(x): # функция линейной зависимости
  return x

lin_x = torch.linspace(-10, 10, 21) # генерируем числа от -10 до 10 в количестве 21
lin_y = [linear(x) for x in lin_x] # высчитываем для каждого входного x значение функции

plt.plot(lin_x, lin_y) # строим график
plt.show()
```

image.png

```python
def sigmoid(x): # нелинейная функция: сигмоида
  return torch.exp(x)/(1 + torch.exp(x))

sig_x = torch.linspace(-10, 10, 21) # генерируем числа от -10 до 10 в количестве 21
sig_y = [sigmoid(x) for x in lin_x] # высчитываем для каждого входного x значение функции

plt.plot(sig_x, sig_y, 'r')
plt.show()
```

> В функцию можно передать больше значений, можно использовать внутри функции другую функцию, можно из них строить сложные цепочки.

> Интересный момент. Если мы измерим с какой скоростью за определённый промежуток времени меняется значение на её выходе, то обнаружим что появляется другая функция. Функция производит другую функцию и имя её ***производная***.



Рассмотрим производную линейной функции.

Мы видим постоянные значения так как первообразная возрастает с постоянной скоростью. Каждый шаг меняет функцию на одну единицу.

```python
derlin_x = torch.linspace(-10, 10, 21, requires_grad=True)
derlin_y = [torch.autograd.grad(linear(x), x)[0] for x in derlin_x]

plt.plot(derlin_x.detach().numpy(), derlin_y, 'ro')
plt.show()
```

Рассмотрим производную сигмоидной функции.

У сигмоиды хорошо видно, что чем круче изгиб первообразной, тем больше значение в производной и наоборот.

```python
dersig_x = torch.linspace(-10, 10, 21, requires_grad=True)
dersig_y = [torch.autograd.grad(sigmoid(x), x)[0] for x in dersig_x]

plt.plot(dersig_x.detach().numpy(), dersig_y, 'ro')
plt.show()
```

Производная показывает как меняется выходное значение родительской функции (первообразной) при поступлении в её вход данных. Чем дальше значение от нуля в производной, тем более крутой подъём или спуск у основной функции.

При помощи производной при обучении нейронных сетей можно объяснять сети где она ошибается при принятии решений и как сильно она должна изменить свой параметр внутри себя, чтобы делать более качественные результаты на выходе. Но об этом подробнее мы будем уже обсуждать на следующих занятиях
