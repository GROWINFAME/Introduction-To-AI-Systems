# **Искусственный интеллект. Машинное обучение. Глубокое обучение**


image.png

1. ***Искусственный интеллект*** является свойством машин, компьютерных программ и систем выполнять интеллектуальные и творческие функции человека, самостоятельно находить способы решения задач, уметь делать выводы и принимать решения.
2. ***Машинное обучение*** является одним из направлений искусственного интеллекта. Основной принцип заключается в том, что машины получают данные и «обучаются» на них. В настоящее время это наиболее перспективный инструмент для бизнеса, основанный на искусственном интеллекте. Системы машинного обучения позволяют применять знания, полученные при обучении, на индетичных данных, которые она видела в процесе обучения. Машинное обучение преуспевает в таких задачах, как распознавание лиц, распознавание речи, распознавание объектов, перевод иностранных языков, и многих других.
3. ***Глубокое обучение*** является подмножеством машинного обучения. Оно использует некоторые методы машинного обучения для решения реальных задач, используя нейронные сети, которые могут имитировать человеческое принятие решений. Процесс обучения называется глубоким, так как структура искусственных нейронных сетей состоит из нескольких входных, выходных и скрытых слоев. Глубокое обучение может быть дорогостоящим и требует огромных массивов данных для обучения. Это объясняется тем, что существует огромное количество параметров, которые необходимо настроить для алгоритмов обучения, чтобы избежать ложных срабатываний. 
  
  Например, алгоритму глубокого обучения может быть дано указание «узнать», как выглядит кошка. Чтобы произвести обучение, потребуется огромное количество изображений для того, чтобы научиться различать мельчайшие детали, которые позволяют отличить кошку от, скажем, гепарда или пантеры, или лисицы.

## **Машинное обучение**
## **Обучение с учителем (Supervised Learning)**
>***Обучение с учителем (Supervised Learning)*** - способ машинного обучения для обучения которого используется обучающий набор данных, содержащий пары «известный вход — известный выход». Известный вход это объекты, описанные набором характерных признаков. Для каждого такого объекта некоторой выборки известен целевой признак (это предварительно размеченные данные человеком-экспертом), который нужно предсказать модели. По-сути это правильный ответ, на который должна опираться модель в процессе обучения. Поэтому этот процесс и называется обучение с учителем.

>К этой категории относятся почти все современные способы применения глубокого обучения, такие как распознавание образов, распознавание речи, классификация изображений и перевод с одного языка на другой.

По сути обучение с учителем это задача следующего вида: 

Имеются

*   множество $𝐗$ признаков характеризующией объект. (прогнозирующая переменная)
*   множество $𝐘$ правильных ответов - значения, которые должна выдавать модель после обучения (целевая или зависимая от $x$ переменная).

Предполагается, что существует некоторая скрытая функциональная зависимость между ними (то есть какая-то закономерность между $X$ и $Y$): 

> $f=𝐗⟶𝐘$

Однако эта зависимость неизвестна. Известна лишь совокупность пар $S(объект, ответ)$, которая называется обучающей выборкой.


> $S=\{(x_i,y_{x_i}=𝒇(x_i))\}$

Требуется найти приближенный вид этой функции $f$ путем построения аппроксимирующей функции $\hat{f}=𝐗⟶𝐘$ с помощью алгоритма машинного обучения. То есть, на основе этих данных требуется восстановить зависимость, построить алгоритм, способный для любого объекта выдать достаточно точный ответ.

> $\hat{f}\approx f(x)$

В результате, обученный алгоритм должен делать обобщение всех изученных наблюдений (признаков) обучающей выборки и найдя определённую зависимость, выдавать правильные результаты на невиданных ранее образцах (тестовая выборка).

Для этого используются специальные методы оптимизиации (например, метод градиентного спуска). Об этом мы будем говорить в последующих занятиях.

####**Типы задач обучения с учителем.**

В общем случае, задачи обучения с учителем делятся на следующие типы :

1. **Классификация (Classification):** геометрический смысл задачи классификации, состоит в построении разделяющей функции между объектами, имеющие разные метки класса. Построив такую функцию, для любого нового объекта, можно будет предсказать к какому классу он относится.

  **Классификация предсказывает метку класса (class label), из предопределенного дискретного множества возможных классов в данной задаче, по набору признаков конкретного объекта.**

    Таким образом в задаче классификации:
  * $X$ - характеристики (признаки), описывающие объекты
  * $Y$ - метки классов для каждого объекта

  Рассмотрим рисунок ниже. Например, каждое электронное письмо с текстом, описывается двумя характеристиками $x_1$ и $x_2$. Каждое из этих писем может быть классифицировано как принадлежащее одному из двух классов: "спам" или "не спам". Для классификации моделью строится наиболее оптимальная разделяющая функция (в данном случае прямая). После обучения и построения такой функции на обучающих данных, теперь для любого нового письма, можно будет предсказать к какому классу он относится.
  
* Задача классификации с двумя классами называется ***бинарной классификацией***. 
  
* Если же в задаче рассматривается более 2 классов, то такая задача называется ***многоклассовой классификацией.***

2. **Регресиия (Regression):** геометрический смысл задачи регрессии, состоит в том, что она пытается описать взаимосвязь (зависимости) между известным $X$ и $Y$ в виде некоторой функции. После её построения, (то есть обучения модели на этих данных) появляется возможность использовать новые $X$ и предсказать для них целевое значение $Y$.

  **Регрессия предсказывает непрерывную выходную переменную - это целое число или число с плавающей точкой.**

    Таким образом в задаче регрессии:
  * $X$ - характеристики (признаки), описывающие объекты
  * $Y$ - непрерывная выходная целевая переменная (некоторая характеристика объектов)

  Рассмотрим рисунок ниже. Например, требуется предсказать цену дома $Y$ на основе набора характеристик района, в котором находится дом (в данном случае имеется один какой-то признак $x$). Сперва, для этого моделью строится оптимальная функция, описывающая взаимосвязь между известным $X$ и целевым $Y$ (в данном случае строится линейная функция). После её построения, (то есть обучения модели на этих данных) появляется возможность использовать новые $X$, которые модель не видела ранее и предсказать для них целевое значение $Y$.
   
   Следует отметить, что целевая переменная представляет собой сумму, которая может быть любым числом в определённом непрерывном диапазоне, например от 10000 до 500000 долларов.

#### **Другие задачи машинного обучения.**
* ***Обучение без учителя (Unsupervised Learning)***  - способ машинного обучения без привлечения каких-либо размеченных данных (без заранее известного выхода) для обучения.
* ***Обучение с частичным привлечением учителя (Semi-Supervised Learning)*** - разновидность обучения с учителем, которое использует обычно небольшое количество размеченных данных (ввиду их ограниченности) и большое количество неразмеченных данных.
* ***Обучение с подкреплением (Reinforcement Learning).*** один из способов машинного обучения, в ходе которого испытуемая система (агент) обучается, взаимодействуя с некоторой средой. Здесь подкрепление обозначает награду или наказание за некоторый получившийся результат, зависящий не только от самих принятых решений агента, но и неподконтрольной окружающей среды. Смысл задачи состоит в максимизации скалярной величины - награды.

# **Тенозоры как способ описания объектов в виде признаков**

В задачах машинного обучения признаки хранятся в виде тензора.

> Как уже было рассмотрено ранее ***тензор*** - это объект линейной алгебры, который линейно преобразует элементы одного линейного пространства в элементы другого.

В частном случае тензор 

1.   Нулевого ранга - скаляр ( $a$ )
2.   Первого ранга - вектор ( $a_i$ )
3.   Второго ранга - матрица ( $a_{ij}$ )
4.   Третьего ранга - многомерная матрица ($a_{ijm}$)

Чтобы было понятнее, перечислим несколько примеров тензоров с данными, которые чаще всего встречаются в задачах машинного обучения. Данные, которыми вам придется манипулировать, почти всегда будут относиться к одной из следующих категорий:


1.   ***Векторные данные*** — двумерные тензоры с формой (образцы, признаки объекта);
2.   ***Временные ряды или последовательности (например, текстовые)*** — трехмерные тензоры с формой (образцы, метки времени, признаки объекта);
3.   ***Изображения*** — четырехмерные тензоры с формой (образцы, высота, ширина, цвет - для Tensorflow) или с формой (образцы, цвет, высота, ширина - для Pytorch).
4.   ***Видео*** — пятимерные тензоры с формой (образцы, кадры, высота, ширина, цвет) или с формой (образцы, кадры, цвет, высота, ширина).

### **Типы признаков**

> ***Признаки объекта*** - набор существенных характеристик определённого объекта. Используя и анализируя их, алгоритмы машинного обучения например пытаются предсказать какую-то характеристику об этом объекте (***регрессия***) или классифицировать объект по определённому типу (***классификация***). Успех обучения алгоритма существенно зависит от качества и значимости признаков.

Признаки могут быть:

* ***вещественными*** (вес - 45.5, рост - 150.5, пиксели и тд). Поскольку в совокупности диапазон или масштаб одной переменной сильно отличается от масштаба другой, это может ухудшить качество обучения моделей. Модель может ошибочно посчитать, что раз признак большой, то он полезнее остальных. Поэтому такие признаки часто подвергаются единому масштабированию. Среди таких методов преобразования часто выделяют ***Min-Max нормализация*** и ***Стандартизация (Z-оценка)*** Подробнее об этих методах написано в задании 1 ниже.
* ***бинарными*** (женщина/мужчина или больной/небольной) Такие признаки перед обучением представляют в виде 0 и 1.
* ***категориальные*** (ряд словесных описаний характеристики). 

  Например цвет автомобиля: (красный, синий, зелёный). Их чаще всего представляют:
  * в виде числовой последовательности (1, 2, 3);
  * увеличив размерность вектора, можно сделать бинарными. Для этого используется one-hot кодирование. 
  
    красный: [1, 0, 0] (красный, не синий, не зелёный), 
    
    синий: [0, 1, 0] (не красный, синий, не зелёный), 
    
    зелёный: [0, 0, 1] (не красный, не синий, зелёный). 
    
    Это всё объединяется в единую матрицу таких векторов.

### **Векторные данные**

> Наиболее часто встречающаяся форма данных. Первая ось — это ось образцов, а вторая — ось признаков. Каждый образец представляет собой вектор с какими-то данными, описывающие существенные признаки.

Например: набор данных для предсказания цены дома (506 образцов, где каждый образец - вектор из 13 признаков). 

Каждая объект данных — это район города Бостон. У нас есть информация об уровне преступности (CRIM), качестве воздуха (NOX), транспортной доступности (RAD), налогах (TAX), количестве учителей (PTRATIO), социальном положении населения (LSTAT) и некоторые другие показатели. Целевой переменной является медианная цена недвижимости в каждом из районов (PRICE). Именно ее и нужно научиться предсказывать в данной задаче.

Таким образом, требуется предсказать цену дома $Y$ на основе набора признаков $X$.






```python
import matplotlib.pyplot as plt
import pandas as pd # для работы с табличными данными
from sklearn.datasets import load_boston # библиотека машинного обучения в котором также есть наборы данных

boston = load_boston() # загрузим набор данных
boston_df = pd.DataFrame(boston['data'] ) # загружаем набор данных в датафрейм
boston_df.columns = boston['feature_names'] # загружаем колонки
boston_df['PRICE']= boston['target'] # загружаем целевую переменную
boston_df # смотрим получившийся датафрейм
```

```python
boston_df.describe() # статистика по данным
```

```python
plot_cols = ['CRIM', 'PRICE'] # Пишем признаки, которые хотим рассмотреть. Например как зависит цена от преступности
plot_features = boston_df[plot_cols] # Вычленяем из датафрейма эти признаки
plt.figure(figsize=(10, 10))
plt.plot(plot_features, 'ro',) # Строим график этих признаков по отдельности
plt.xlabel(plot_cols[0])
plt.ylabel(plot_cols[1])
```

### **Временные ряды или последовательности**

> В таких данных важную роль играет ось времени, поэтому данные хранятся в трёхмерном тензоре. Первая ось — это ось образцов, второй всегда является осью времени, а третью осью является ось признаков.

Например: набор данных для прогноза погоды. Имеет 70091 образцов. Каждый образец содержит 14 различных признаков, таких как температура воздуха, атмосферное давление и влажность. Они собирались каждые 10 минут, начиная с 2003 года. 

Таким образом, требуется предсказать погоду в градусах $Y$ на основе 14 признаков $X$.

В данном случае рассматриваются наблюдения с 2009 до 2016 года, не каждые 10 минут, а каждый час.

Даже при том, что черно-белые изображения имеют только один канал цвета и могли бы храниться в двумерных тензорах, по соглашениям тензоры с изображениями всегда имеют три измерения, где для черно-белых изображений отводится только один канал цвета.

```python
# Пример RGB изображений цветов
import torchvision
import torch
import matplotlib.pyplot as plt

CIFAR_train = torchvision.datasets.CIFAR10('./', download=True, train=True)
train_data = torch.tensor(CIFAR_train.data, dtype=torch.float32)
print("Размерность набора данных: ", train_data.data.shape)
X_train = train_data.permute(0, 3, 1, 2) # признаки
print("Правильная размерность для Pytorch:", X_train.shape)
classes = CIFAR_train.classes # список классов
Y_train = torch.tensor(CIFAR_train.targets) # целевая переменная Y для каждой картинки в данных. Цифра по сути это индекс класса в списке classes.
print("Список классов: ", classes)

plt.figure(figsize=(20,2))
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.title(classes[Y_train[i]])
    plt.imshow(train_data[i]/255)
```

### **Видео**

> Видеоданные — один из немногих типов данных, для хранения которых требуются пятимерные тензоры. Видео можно представить как последовательность кадров, где каждый кадр — цветное изображение. Каждый кадр можно сохранить в трехмерном тензоре (высота, ширина, цвет), соответственно, их последовательность можно сохранить в четырехмерном тензоре (кадры, высота, ширина, цвет), а целый пакет разных видеороликов — в пятимерном тензоре с формой (образцы, кадры, высота, ширина, цвет).


```python
import cv2
import torch
import urllib
import numpy as np
import matplotlib.pyplot as plt

# СКАЧИВАНИЕ ВИДЕО
url_link = "https://dl.fbaipublicfiles.com/pytorchvideo/projects/archery.mp4" # ссылка на видео
video_path = 'archery.mp4'
try: urllib.URLopener().retrieve(url_link, video_path)
except: urllib.request.urlretrieve(url_link, video_path)

frames = [] # список кадров нашего видео

stream = cv2.VideoCapture(video_path) # фиксатор кадров от OpenCV

while(True): # поточное циклическое считывание кадров видео
  (grabbed, frame) = stream.read()
  if(not grabbed):
    break
  frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) # поскольку OpenCV работает с BGR, то переводим в RGB
  frames.append(frame) # добавляем каждый кадр в наш список

video_tensor = torch.tensor(frames) # преобразуем наш список в тензор от pytorch
video_tensor = video_tensor[None, :, :, :, :] # Добавляем ещё одно измерение. Теперь у нас (образцы, кадры, высота, ширина, цвет)
video_tensor = video_tensor.permute(0, 1, 4, 2, 3) # Поскольку в Pytorch цвет стоит первее высоты и ширины, то меняем местами
print("Размерность одного видео: ", video_tensor.shape) # теперь (образцы, кадры, цвет, высота, ширина)

plt.imshow(frames[0])
```

# **Задача 1. Масштабирование признаков (Feature scaling).**

Реализовать:

*   ***Стандартизацию (Z-оценка)*** для векторных данных Boston
*   ***Min-Max нормализацию*** для изображений из CIFAR10.

***Стандартизация и нормализация*** - это два различных способа масштабирования данных. Как уже было сказано ранее, так случается, что диапазон одной переменной признака может сильно отличаться от диапазона другой и это может ухудшить качество обучения многих моделей машинного обучения (в частности нейронных сетей). Модель может ошибочно посчитать, что раз признак большой, то он полезнее остальных. 

Например, у нас есть переменные возраст, зарплата и рост. Их диапазоны могут варьироваться как: (18–100 лет), (25,000–75,000 рублей) и (1–2 метра) соответственно.

Поэтому перед тем как обучать модель часто применяется масштабирование признаков.

---

***Стандартизация*** изменяет масштаб индивидуально для каждого признака набора данных, чтобы среднее значение μ (мат.ожидание) равнялось 0, а стандартное отклонение σ равнялось 1. Здесь нет конкретной верхней или нижней границы для максимального и минимального значений. Для этого используется следующая формула:

$X_{stand}=(X-μ)/σ$

Результатом стандартизации является то, что признаки будут масштабироваться так, чтобы они имели свойства стандартного нормального распределения.

image.png

---

***Нормализация*** изменяет масштаб индивидуально для каждого признака набора данных так, чтобы каждое значение находилось между 0 и 1. Для этого используется следующая формула:

$X_{norm} = (X- min(X))/(max(X)-min(X))$

где $min(X)$ - минимальное значение в конкретном признаке

$max(X)$ - максимальное значение в конкретном признаке


---

**Стандартизация и нормализация?**

К сожалению, единого мнения или правила не вырабатно по этому вопросу, однако есть некоторые рекомендации.

> * Нормализация полезна, когда распределение данных не соответствует распределению Гаусса. Это может быть полезно в алгоритмах, которые не принимают в своё внимание распределения данных, таких как нейронные сети или метод k-ближайших соседей. **В нейронных сетях нормализация является важным этапом предварительной обработки.**
* Стандартизация может быть очень полезна, когда данные следуют распределению Гаусса для алгоритмов учитывающие распределение данных. Хотя это и не обязательно. Поскольку стандартизация не имеет ограничивающего диапазона, поэтому даже если в данных есть выбросы, они не повлияют на результат масштабирования, в отличии от нормализации.

В целом, нам ничего не мешает сначала отдельно обучить модель на сырых данных, потом на нормализованных и затем на стандартизированных. После этого мы можем сравнить и выбрать лучший вариант масштабирования.

Однако следует иметь ввиду, что не все алгоритмы требуют масштабирования данных, поэтому это следует проверять перед их использованием.
